```python
fig, ax = plt.subplots(figsize=(14, 8))
plt.plot(df_pca["principal_component"], df_pca["explained_variance"], marker="o", label="Individual Explained Variance")
plt.plot(df_pca["principal_component"], df_pca["cumsum_explained_variance"], marker="o", label="Cumulative Explained Variance")
plt.plot(df_pca["principal_component"], [0.95] * len(df_pca), color="black", linewidth=2, linestyle="--", label="95% Explained Variance")
ax.set_xticklabels(["{:,.0f}".format(x) for x in ax.get_xticks()])
ax.set_yticklabels(["{:,.0%}".format(x) for x in ax.get_yticks()])
plt.title("PCA Explained Variance", fontsize=20, fontweight="bold")
ax.title.set_position([.5, 1.025])
plt.xticks(range(0,70), range(0,70), fontsize=8, fontweight="bold")
plt.yticks(fontsize=12, fontweight="bold")
plt.xlabel("Principal Component Number", fontsize=16, fontweight="bold")
plt.ylabel("% of Explained Variance", fontsize=16, fontweight="bold")
plt.grid(False)
plt.legend(loc="best", frameon=False)
plt.tight_layout()
plt.show()
```

### KMeans

Wybór liczby klastrów

```python
def review_model(model, df):
    y = model.fit_predict(df)
    show_groups_counts(y)
    show_metrics_scores(df, y)
    
def show_groups_counts(colors):
    unique, counts = np.unique(colors, return_counts=True)
    print("Group - Count")
    for k in range(len(unique)):
        print(f"{unique[k]} - {counts[k]}")


def show_metrics_scores(X, y):
    print(f"Metryka Silhouette'a: {silhouette_score(X, y)}")
    print(f"Metryka Calinskiego-Harabasza: {calinski_harabasz_score(X, y)}")
    print(f"Metryka Daviesa-Bouldina: {davies_bouldin_score(X, y)}")
```

```python
def plot_scores(estimator, X, max_clusters=20):
    
    cluster_num_seq = range(2, max_clusters+1)
    
    scores = [silhouette_score, calinski_harabasz_score, davies_bouldin_score]
    
    scores_values = count_clustering_scores(X, cluster_num_seq, estimator, scores)
    
    plt.plot(cluster_num_seq, scores_values[silhouette_score], 'bx-')
    plt.xlabel('Liczba klastrów')
    plt.ylabel('Miara metryki silhouette')
    plt.title('Wartość miary silhouette')
    plt.show()
        
    plt.plot(cluster_num_seq, scores_values[calinski_harabasz_score], 'bx-')
    plt.xlabel('Liczba klastrów')
    plt.ylabel('Miara metryki Calinski-Harabasz')
    plt.title('Wartość miary Calinski-Harabasz')
    plt.show()
    
    plt.plot(cluster_num_seq, scores_values[davies_bouldin_score], 'bx-')
    plt.xlabel('Liczba klastrów')
    plt.ylabel('Miara metryki Davies-Bouldin')
    plt.title('Wartość miary Davies-Bouldin')
    plt.show()
    
    
def count_clustering_scores(X, cluster_num, model, scores_fun_list):
    if isinstance(cluster_num, int):
        cluster_num_iter = [cluster_num]
    else:
        cluster_num_iter = cluster_num
        
    scores = {}    
    for x in scores_fun_list:
        scores[x] = []
        
    for k in cluster_num_iter:
        if type(model).__name__[-15:] == 'GaussianMixture':
            model.set_params(n_components=k)
        else:
            model.set_params(n_clusters=k)
      
        labels = model.fit_predict(X)
        for a in scores_fun_list:
            temp = scores[a]
            temp.append(a(X, labels))
            scores[a] = temp
    
    if isinstance(cluster_num, int):
        return scores[0]
    else:
        return scores
```

```python
def plot_inertias_for_KMean(K):
    inertias = []
    for k in K:
        km = KMeans(n_clusters=k)
        km = km.fit(df_array_reduced)
        inertias.append(km.inertia_)
    plt.plot(K, inertias, 'bx-')
    plt.xlabel('k')
    plt.ylabel('Inertia')
    plt.title('Elbow Method For Optimal k')
    plt.show()
```

### Analiza modeli wyżej wymienionych tj. KMeans dla odpowiedniej liczby klastrów: 3, 10, 20

Poniżej bedziemy mogli zobaczyc licznosci poszczegolnych grup jak i uzyskane metryki przez nasze algorytmy. Zasotosowane metryki to odpowiednio:
- Metryka Silhouette'a
- Metryka Calinskiego-Harabasza
- Metryka Daviesa-Bouldina

Interpretowalność powyższych metryk:

*Silhouette metric* opisuje skuteczność klasteryzacji. Najlepsza wartość to 1, a najgorsza to -1. Wartości bliskie 0 wskazują na nakładające się klastry. Wartości ujemne ogólnie wskazują, że próbka została przypisana do niewłaściwego klastra, ponieważ inny klaster jest bardziej podobny.

*Calsinski Harabasz* znana również jako Variance Ratio Criterion jest metryką określającą stosunek rozczucenia w obrębie klastra a rozrzuceniem między klastrami. Klasteryzacja jest tym lepsza im wyższa jest wartość tej metryki

*Davies-Bouldin* średnia miara podobieństwa, gdzie podobieństwo jest stosunkiem odległości wewnątrz skupień do odległości między skupieniami. Lepszy wynik uzyskają więc klastry bardziej oddalone od siebie i mniej rozproszone. Minimalny wynik to 0 i im mniejsza wartość tej metryki tym lepsze jest nasze klastrowanie

# DBSCAN

Szczerze mówiąc nie do końca chyba ogarniam ten DBSCAN i te parametry ustawiane jesli ogarniesz to mozesz jakos ladniej to opisac bo ja bazuje na jakims artykule naukowym

```python
from sklearn.cluster import DBSCAN
```

```python
fig = plt.figure(figsize=(20, 10))
fig.subplots_adjust(hspace=.5, wspace=.2)
i = 1
for x in range(10, 0, -1):
    eps = 3+x
    db = DBSCAN(eps=eps, min_samples=76).fit(df_array_subset)
    labels_subset = db.fit_predict(df_array_subset)
    print(eps)
    ax = fig.add_subplot(2, 5, i)
    ax.text(1, 4, "eps = {}".format(round(eps, 2)), fontsize=25, ha="center")
    plt.scatter(x=tsne_projection_2[:,0], y = tsne_projection_2[:,1], alpha=0.4, c = labels_subset, cmap='tab20')
    
    i += 1
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
nearest_neighbors = NearestNeighbors(n_neighbors=76)
neighbors = nearest_neighbors.fit(df_array_subset)
distances, indices = neighbors.kneighbors(df_array_subset)
distances = np.sort(distances[:,10], axis=0)
fig = plt.figure(figsize=(5, 5))
plt.plot(distances)
plt.xlabel("Points")
plt.ylabel("Distance")
```

Metodą łokcia możemy zauważyć, że odpowiednim eps wydają się wartości pomiędzy 4-6
